```python
from pymongo import MongoClient
import dateutil
from math import cos, asin, sqrt
import math
import datetime
from geopy.distance import great_circle as gc
import matplotlib.pyplot as plt
import numpy as np
import pymc
import pymc.Matplot as pm
```

```python
def seconds(x):
    t=x['ts']
    return datetime.timedelta(days=t.day,hours=t.hour,
                              minutes=t.minute,seconds=t.second).total_seconds()
```

```python
dateb = dateutil.parser.parse("2016-09-13T02:00:00.000-03:00")
datee = dateutil.parser.parse("2016-09-14T02:00:00.000-03:00")

carro='52345'
#52345 52301 52310 52338 52299
```

```python
##
client = MongoClient()
db = client['msgstore']
col = db['messages']
y = col.find({
        "raw.trip_id":  {'$in': ['477P-10-0','477P-10-1']},
        "raw.vs.p":carro,
        'ts': {
                "$gte": dateb,
                "$lte": datee
              }
        })
```

```python
x = y.next()
c=[dictio.values()[1] for dictio in x['raw']['vs']].index(carro)
t0 = seconds(x)
#carro = x['raw']['vs'][c]['p']
pxini = x['raw']['vs'][c]['px']
pyini = x['raw']['vs'][c]['py']
```

```python
T=[]
D=[]
V=[]
X=[]
X.append(x)
```

```python
for i in range(y.count()-1) :
    x2 = y.next()
    #l = len(x2['raw']['vs'])
    #if c<l:
    c=[dictio.values()[1] for dictio in x2['raw']['vs']].index(carro)    
    #if x2['raw']['vs'][c]['p']==carro:
    X.append(x2)
    V.append(x2['raw']['trip_id'][-1])
    t1 = seconds(x2)
    t = t1-t0
    T.append(t/3600)
    pxf = x2['raw']['vs'][c]['px']
    pyf = x2['raw']['vs'][c]['py']
    dist = gc((pxini,pyini), (pxf, pyf)).km
    D.append(dist)
```

```python
T = np.array(T)
D = np.array(D)
V = np.array(V)    
plt.plot(T[V=='0'], D[V=='0'], '.r',T[V=='1'], D[V=='1'], '.b')
plt.legend(['Ida','Volta'])  
plt.xlabel('Tempo (h)')
plt.ylabel('Distancia (Km)')
plt.title('Linha 477P-10 | Carro: '+carro)
plt.show()
```

```python
x = T
f = D
```

## Linear Regression
### Linear Basis Function

$$\begin{align*}
y(\mathbf{x},\mathbf{w}) =& w_{0} + \sum_{j=1}^{M-1} w_{j}\phi_{j}(x)\\
y(\mathbf{x},\mathbf{w}) =& \sum_{j=0}^{M-1} w_{j}\phi_{j}(x),\quad \phi_{0}=1
\end{align*}$$

$$\phi_{j}(x)= exp \bigg(\frac{-(x - \mu_{j})^{2}}{2s^{2}} \bigg)$$

### Bayesian Linear Regression
We want to estimate the uncertainty on a set of parameters $\mathbf{w}$ given observed data $t$.

Bayes Theorem:
$$p(\mathbf{w}|t) \propto p(t|\mathbf{w})p(\mathbf{w})$$

Parameter Prior Distribution:
$$p(\mathbf{w}|\mu_{0}, \sigma_{0}^{2}) = \mathcal{N}(\mathbf{w}|\mu_{0}, \sigma_{0}^{2})$$

Conjugate Prior of a Normal Likelihood function with known variance/precision: Normal as well.

```python
def model(x, f): 
    #parameter priors
    mu1 = pymc.Normal('mu1', 10, .1)
    mu2 = pymc.Normal('mu2', 10, .1)
    mu3 = pymc.Normal('mu3', 10, .1)
    size1 = pymc.Normal('size1', 0.1, 10)
    size2 = pymc.Normal('size2', 0.1, 10)
    size3 = pymc.Normal('size3', 0.1, 10)
    w1 = pymc.Normal('w1', 20, 10)
    w2 = pymc.Normal('w2', 20, 10)
    w3 = pymc.Normal('w3', 20, 10)
    w0 = pymc.Normal('w0', 0.1, 10)

    

    @pymc.deterministic(plot=False)
    def gauss(w0=w0, w1=w1, w2=w2, w3=w3, x=x, 
              size1=size1, size2=size2, size3=size3, mu1=mu1,mu2=mu2,mu3=mu3):
        g1=w1*np.exp(-np.power(x - mu1, 2.) / (2 * size1**2))
        g2=w2*np.exp(-np.power(x - mu2, 2.) / (2 * size2**2))
        g3=w3*np.exp(-np.power(x - mu3, 2.) / (2 * size3**2))
        return g1+g2+g3+w0
    #likelihood function
    y = pymc.Normal('y', mu=gauss, tau=.1, value=f, observed=True)
    return locals()
```

```python
MDL = pymc.MCMC(model(x,f))
MDL.sample(10000, 1000)
```

```python
pm.figure(figsize=(15,8))
pm.histogram(MDL.trace('w1')[:], 'w1',rows=1,columns=3,num=1,verbose=0)
pm.histogram(MDL.trace('w2')[:], 'w2',rows=1,columns=3,num=2,verbose=0)
pm.histogram(MDL.trace('w3')[:], 'w3',rows=1,columns=3,num=3,verbose=0)
plt.tight_layout()
plt.show()
```

```python
pm.figure(figsize=(15,8))
pm.histogram(MDL.trace('mu1')[:], 'mu1',rows=1,columns=3,num=1,verbose=0)
pm.histogram(MDL.trace('mu2')[:], 'mu2',rows=1,columns=3,num=2,verbose=0)
pm.histogram(MDL.trace('mu3')[:], 'mu3',rows=1,columns=3,num=3,verbose=0)
plt.tight_layout()
plt.show()
```

```python
pm.figure(figsize=(15,8))
pm.histogram(MDL.trace('size1')[:], 'size1',rows=1,columns=3,num=1,verbose=0)
pm.histogram(MDL.trace('size2')[:], 'size2',rows=1,columns=3,num=2,verbose=0)
pm.histogram(MDL.trace('size3')[:], 'size3',rows=1,columns=3,num=3,verbose=0)
plt.tight_layout()
plt.show()
```

```python
# extract and plot results
y_min = MDL.stats()['gauss']['quantiles'][2.5]
y_max = MDL.stats()['gauss']['quantiles'][97.5]
y_fit = MDL.stats()['gauss']['mean']
plt.plot(x,f,'b', marker='None', ls='-', lw=1, label='True')
#plt.errorbar(x,f,yerr=f_error, color='r', marker='.', ls='None', label='Observed')
plt.plot(x,y_fit,'k', label='Fit')
plt.fill_between(x, y_min, y_max, color='0.5', alpha=0.5)
plt.ylabel('Distance (Km)')
plt.xlabel('Time (h)')
plt.legend()
plt.show()
```
