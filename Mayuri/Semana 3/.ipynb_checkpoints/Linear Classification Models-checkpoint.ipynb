```python
import csv
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn import linear_model
%matplotlib inline
from sklearn.metrics import classification_report
```

```python
file1 = open('2017.csv', "rb")
reader = csv.reader(file1, delimiter=';')
features = list(reader)
file1.close()
features = features[:1][0]

file3 = open('d2017.csv', "rb")
reader3 = csv.reader(file3, delimiter=';')
data3 = list(reader3)
file3.close()

d3 = np.array(data3, dtype=np.float64)
T = pd.DataFrame(d3[:, :])
T.columns = features[2:]
X = np.concatenate((d3[:70, 2:-1],d3[85:, 2:-1]))
y = np.concatenate((d3[:70, -1],d3[85:, -1])) #targets

testx = d3[70:85, 2:-1]
testy = d3[70:85, -1]
```

```python
glass = pd.read_csv('glass.csv',sep=';')

traindata = pd.concat([glass.iloc[:50, :-1], glass.iloc[55:100, :-1],
                      glass.iloc[105:150, :-1], glass.iloc[155:170, :-1],
                      glass.iloc[173:180, :-1],glass.iloc[181:209, :-1]])

trainclas = pd.concat([glass.iloc[:50, -1], glass.iloc[55:100, -1], 
                      glass.iloc[105:150, -1], glass.iloc[155:170, -1],
                      glass.iloc[173:180, -1],glass.iloc[181:209, -1]])

testdata  = pd.concat([glass.iloc[50:55, :-1],glass.iloc[100:105, :-1],
                      glass.iloc[150:155, :-1],glass.iloc[170:173, :-1],
                      glass.iloc[180:181, :-1],glass.iloc[209:, :-1]])

testclas  = pd.concat([glass.iloc[50:55, -1],glass.iloc[100:105, -1],
                      glass.iloc[150:155, -1],glass.iloc[170:173, -1],
                      glass.iloc[180:181, -1],glass.iloc[209:, -1]])


Xg = traindata.as_matrix()
yg = trainclas.as_matrix()
testxg = testdata.as_matrix()
testyg = testclas.as_matrix()
```

## Logistic Regression

## $$p(C_{1}\mid\mathbf{\phi})=y(\mathbf{\phi}) = \sigma (\mathbf{w}^T \mathbf{\phi})$$
## $$\sigma(a) = \frac{1}{1+exp(-a)}$$

Bishop describes a Maximum Likelihood method to estimate $\mathbf{w}$. The implementation minimizes a cost function with L2 or L1 regularization (http://scikit-learn.org/stable/modules/linear_model.html#logistic-regression).


```python
logreg = linear_model.LogisticRegression(C=1,penalty='l2', max_iter=1000, solver='saga',verbose=0)
logreg.fit(X, y)
tr1 = sum(y == logreg.predict(X))
te1 = sum(testy == logreg.predict(testx))
```

## Perceptron

## $$y(\mathbf{x})=f(\mathbf{w}^{T}\mathbf{\phi(x)})$$

To estimate the parameters, we consider the $perceptron\; criteria$
## $$E_{p}(\mathbf{w})= -\sum_{n \in \mathcal{M}}\mathbf{w}^{T}\mathbf{\phi}_{n}t_{n}$$

and use the stochastic gradient descent algorithm, leading to:

## $$\mathbf{w}^{\tau +1} = \mathbf{w}^{\tau} - \eta\nabla E_{p}(\mathbf{w}) = \mathbf{w}^{\tau} + \eta\phi_{n}t_{n}$$

In the scikit-learn implementation there isn't a learning rate parameter nor is possible to change the activation function.

```python
percep = linear_model.Perceptron(max_iter=1000, tol=None)
percep.fit(X,y)
tr2 = sum(y == percep.predict(X))
te2 = sum(testy == percep.predict(testx))
```

## Least Squares for Classification

Each Class $C_{k}$ is described by its own linear model:

## $$y_{k}(\mathbf{x})= \mathbf{w}_{k}^{T}\mathbf{x} + w_{k0} = \mathbf{W}^{T}\mathbf{X}$$

To determine $\mathbf{W}$, we minimize the Residual Sum of Squares function. The implmentation also includes a penalty on the parameter size:

## $$\min_{w}\|\mathbf{XW}-\mathbf{T}\|^{2}-\alpha \|\mathbf{W}\|^2 $$

```python
ll = linear_model.RidgeClassifier(solver='lsqr')
ll.fit(X,y)
tr3 = sum(y == ll.predict(X))
te3 = sum(testy == ll.predict(testx))
```

```python
logreg2 = linear_model.LogisticRegression(C=0.1,penalty='l2', max_iter=1000, solver='sag',verbose=0)
logreg2.fit(Xg, yg)
tr11= sum(yg == logreg2.predict(Xg))
te11= sum(testyg == logreg2.predict(testxg))
```

```python
percep2 = linear_model.Perceptron(max_iter=1000, tol=None,verbose=0)
percep2.fit(Xg,yg)
tr22 = sum(yg == percep2.predict(Xg))
te22 = sum(testyg == percep2.predict(testxg))
```

```python
ll2 = linear_model.RidgeClassifier(solver='lsqr')
ll2.fit(Xg,yg)
tr33 = sum(yg == ll2.predict(Xg))
te33 = sum(testyg == ll2.predict(testxg))
```

## Happiness Dataset Classification: 

```python
target_names = ['less happy','more happy']
print "Logistic Regression:"
print 'Training Hits: '+ str(tr1)+'/'+str(len(X))+' = '+str(float(tr1)/len(X))
print(classification_report(y, logreg.predict(X), target_names=target_names))
print 'Test Hits: '+ str(te1)+'/'+str(len(testx))+' = '+str(float(te1)/len(testx))
print(classification_report(testy, logreg.predict(testx), target_names=target_names))
```

```python
print "Perceptron:"
print 'Training Hits: '+ str(tr2)+'/'+str(len(X))+' = '+str(float(tr2)/len(X))
print(classification_report(y, percep.predict(X), target_names=target_names))
print 'Test Hits: '+ str(te2)+'/'+str(len(testx))+' = '+str(float(te2)/len(testx))
print(classification_report(testy, percep.predict(testx), target_names=target_names))
```

```python
print "Least Squares:"
print 'Training Hits: '+ str(tr3)+'/'+str(len(X))+' = '+str(float(tr3)/len(X))
print(classification_report(y, ll.predict(X), target_names=target_names))
print 'Test Hits: '+ str(te3)+'/'+str(len(testx))+' = '+str(float(te3)/len(testx))
print(classification_report(testy, ll.predict(testx), target_names=target_names))
```
