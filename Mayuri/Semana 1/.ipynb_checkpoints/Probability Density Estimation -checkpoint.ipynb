# Probability Density Estimation

In this notebook, we'll explain two methods of probability density estimation, namely Gaussian Kernel Density Estimation and K-nearest neighbors Density Estimation. These methods will be applied to two unidimensional real datasets with unknown density distributions, explained below. 


```python
import csv
import numpy as np
import heapq
import scipy.stats as stats
import matplotlib.pyplot as plt
```

Countries must be assessed by at least three sources to appear in the CPI. The surveys/assessments are either business people opinion surveys or performance assessments from a group of analysts. ## Histogram Estimator

Standard histogram procedure divides the data in a number of bins with $\Delta$ width and counts the number of ocurrences in each bin. To aproximate the histogram to a density estimator, just divide the frequency (ie, the number of ocurrences) in each bin by the data size times the bin width:

$$p_{i} = \frac{n_{i}}{N\Delta_{i}}$$



## Kernel Density Estimator

Suppose we take $N$ observations from an unknown probability distribution $p(x)$. Also, consider a volume $V$ in a D-dimensional space, small enough such that $p(x)$ inside de volume is constant. The probability of $K$ numbers fall inside this volume V can be modeled by a binomial distribution. Taking the values of mean and variance of this distribution and doing some math, we get to an approximation of $p(x)$:

$$p(x) = \frac{K}{NV}$$ 

If we take the function $k(\mathbf{u})$, given by:

$$k(\mathbf{u}) = \bigg\{ \begin{array}{cccc} 
1,& &    |u_{i}|\leq 1/2 & & i=1,...,D\\
0,& &    otherwise & & \end{array}$$

with $\mathbf{u}$ being the distance between the volume center and a point in the dataset, then we can sum over every point in the dataset N

$$K = \sum_{n=1}^{N} k\bigg(\frac{x - x_{n}}{h}\bigg)$$

and get the total number of K points inside de volume V. Now, substituting K in $p(x) = \frac{K}{NV}$, we have a retangular kernel density estimator:

$$p(x) = \frac{1}{Nh^{D}} \sum_{n=1}^{N} k\bigg(\frac{x - x_{n}}{h}\bigg)$$

where $h^{D}$ is the volume. In fact, $k(\mathbf{u})$ is called a kernel function and can be substituted by many others functions, such as the gaussian:

$$p(x)=\frac1N \sum_{n=1}^{N} \frac{1}{\sqrt{2\pi h^{2}}} exp\bigg\{ -\frac{||x - x_{n}||^{2}}{2h^{2}} \bigg\}$$

which give us the gaussian kernel density estimator (implemented below).


```python
def gau_kernel_estim ( points , data, h ):
    v=[]  
    for po in points:
        N = len(data) #
        xn = np.array(data)
        h2 = np.square(h)
        const = 1/(N*np.sqrt(2*np.pi*h2))
        p = const*np.sum(np.exp(-np.square(po - xn)/(2*h2)))        
        v.append(p)
    return np.array(v);
```

## K-Nearest Neighbors Density Estimator

In the kernel density estimation, we had the function

$$p(x) = \frac{K}{NV}$$,

in which we fixed the volume V and estimated K from the data. 

Another approach is fix K and calculate V from the data, in such a way that the K nearest points to the center of the volume will be inside the volume.

```python
def knn_estim ( points , data, k):
    vec=[]
    for po in points:
        N = len(data)
        xn = np.array(data)
        dist = np.abs(po - xn)
        kp = k+1
        near = heapq.nsmallest(kp, dist)
        r = near[-1]      
        p = k/(N*2*r)
        vec.append(p)
    return np.array(vec);
```

Now we call the density estimation functions:

```python
pdgau1 = gau_kernel_estim(d1, d1, 1)
pdknn1 = knn_estim(d1 , d1, 10)
pdgau3 = gau_kernel_estim(d3, d3, 0.1)
pdknn3 = knn_estim(d3 , d3, 10)
```

We also use a built in scipy function for gaussian kernel density estimation for comparison:

```python
x_pdf1 = stats.kde.gaussian_kde(d1)
x_pdf3 = stats.kde.gaussian_kde(d3)
```

And we plot the results.

```python
plt.figure(1, figsize=(16, 6))
plt.subplot(121)
plt.plot(np.mean(d1), 0.0,'ro')
a1=plt.hist(d1, 20, normed=True)
a2=plt.plot(d1, x_pdf1.evaluate(d1), d1, pdgau1, 'y')
a3=plt.plot(d1, pdknn1, 'k')
plt.xlabel('Corruption Perception Index')
plt.ylabel('p(x)')
plt.title('Density Estimation')
plt.legend(['Mean', 'Python Estimator','Gauss Estimator','KNN Estimator','Norm Histogram'])

plt.subplot(122)
plt.plot(np.mean(d3), 0.,'ro')
plt.hist(d3, 20, normed=True)
plt.plot(d3, x_pdf3.evaluate(d3), d3, pdgau3,'y')
plt.plot(d3, pdknn3, 'k')
plt.xlabel('World Happiness Score')
plt.ylabel('p(x)')
plt.title('Density Estimation')
plt.legend(['Mean','Python Estimator','Gauss Estimator','KNN Estimator','Norm Histogram'])
#plt.tight_layout()
plt.show()
```
